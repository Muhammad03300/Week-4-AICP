```python
# Name: Abubakar
# Task : 2 ( Week 4)
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("menu.csv")

numeric_cols = df.select_dtypes(include=['float64', 'int64'])

corr = numeric_cols.corr()

plt.matshow(corr)
plt.title("Correlation Matrix")
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.colorbar()
plt.show()
```

```python
# Name: Abubakar
# Task : 4 ( Week 4)
import pandas as pd

df = pd.read_csv("menu.csv")

attributes = ['Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein', 
              'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)', 
              'Iron (% Daily Value)']

max_items = {}

for attribute in attributes:
    max_index = df[attribute].idxmax()
    max_item_name = df.loc[max_index, 'Item']
    max_items[attribute] = {'Item': max_item_name, 'Value': df.loc[max_index, attribute]}

max_items_df = pd.DataFrame(max_items)

print("Item(s) with the highest quantity for each attribute:")
print(max_items_df)

summary_link = "https://www.example.com/summary"
print(f"Summary link: {summary_link}")
```

```python
# Name: Abubakar
# Task : 5 ( Week 4)
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("menu.csv")

categories = df['Category'].unique()

for category in categories:
    category_df = df[df['Category'] == category]
    
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Calories', y='Item', data=category_df)
    plt.title(f'Calories in {category}')
    plt.xlabel('Calories')
    plt.ylabel('Item')
    plt.show()
```

```python
# Name: Abubakar
# Task : 6 ( Week 4)
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("menu.csv")

attributes = ['Calories', 'Total Fat', 'Carbohydrates', 'Dietary Fiber', 'Sugars', 'Protein', 
              'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)', 'Calcium (% Daily Value)', 
              'Iron (% Daily Value)']

for attribute in attributes:
    plt.figure(figsize=(10, 6))
    sns.stripplot(x='Category', y=attribute, data=df, jitter=True)
    plt.title(f'Stripplot of {attribute} by Category')
    plt.xlabel('Category')
    plt.ylabel(attribute)
    plt.xticks(rotation=90)
    plt.show()
```
